```python
#without ConfigBox

dict_info = {"name" : "farouk" , "lastname":"elmak"}

dict_info['name']
```

```python
dict_info.name
```

```python
# with ConfigBox


from box import ConfigBox
dict_info = ConfigBox({"name" : "farouk" , "lastname":"elmak"})

dict_info.name

```

```python
# without ensure annotations

def multiplication(x: int , y:int)-> int :
    return x*y
```

```python
multiplication(2,3)
```

```python
multiplication(2,"3")
```

```python
# with ensure annotations


from ensure import ensure_annotations

@ensure_annotations
def multiplication(x: int , y:int)-> int :
    return x*y

multiplication(2,"3") 
```
